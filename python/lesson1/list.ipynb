# Список - `[list]` - он же массив

Список -- изменяемый итерируемый(не один объект, а связка объектов, с которыми можно взаимодействовать) тип данных. 

## Cвойства

- Изменяемый
- Индексируемый
- Хранит любой тип данных

## Синтаксис: 

[ ,x1, x2, x3, ..., xN] -- при этом x может быть любым объектом, любого типа данных.

## Индексация




```python
lst = []
lst2 = [2, 3, 4]
lst3 = [2, "string", 56.65, 'string2']
```

```python
print(lst2[1])
print(lst3[:2])
print(lst3[2:])
print(lst3[::2])
```

# Методы списков

`[list].append(<то, что добавляем>)` - добавление в конце списка к которому обращаемся

```python
print(lst)
lst.append(1)
lst.append(1)
lst.append(2)
lst.append(2)
lst.append(2)
print(lst)
```

`[list].insert(<индекс куда всталяем значение>, <само значение\объект>)` - вставляет на определенную позицицию значение или объет

```python
print(lst)
lst.insert(2, 3) #[1, 1, 3, 2, 2, 2]
lst.insert(2, 3) #[1, 1, 3, 3, 2, 2, 2]
print(lst)
```

`[list].extend(<[list2] второй список>)` - объединяет 2 списка в один общий, добавляя в конец обращаемого списка, добавляемый(в аргументе)

```python

print(lst.extend(lst2)) #при попытке вывода метода мы получаем None, потому что метод ничего не выводитю, так как он занимается объединением списков
print(lst)
```

`[list].remove(<значение\объект>)` - удаляем объект\значение из списка, но если их несколько, то только самое крайнее к левой границе

```python
print(lst)
print(lst.remove(2)) # None, потому что remove ничего не возвращает
print(lst)
```

`[list].pop(<Индекс элемента>)` -- "выдергивает" элемент(объект\значение) из списка, сохраняя его где-либо(по необходимости) 

```python
print(lst)
print(lst.pop(-2)) # забираем 3 но никуда не сохраняем, а просто выводим, доказав этим что значение сохраняется а не удаляется навсегда
print(lst)
val = lst.pop(-2) # забираем значение 2  из списка и сохраняем его в переменной val
print(f'{lst=}, {val=}')
lst3.append(lst.pop(-2)) # забираем значение 2  из списка и сохраняем добавим его в конец lst3
print(f'{lst=}, {lst3=}')
```

`[list].count(<объект\значение>)` -- ищет количество совпадающих с `<значение\объект>` элементов внутри обращаемого списка `[list]`

```python
print(lst.count(1)) # если элемент есть в списке то выведится его количесто > 0
c = lst.count(3)
print(f'{c=}')
print(lst.count(0)) # если искомого элемента нет в списке, получим 0
```

`[list].index(<объект\значение>)` - выводит индекс искомого `<объект\значение>` крайнего слева или ошибку `ValueError` если `<объект\значение>` в списке `[list]` нет

```python
print(lst, lst3)
print(lst.index(4)) # ищем индекс конкретного элемента - 4
print(lst3.index('string')) # ищем индекс конкретного элемента - string
finding_index = lst.index(2)
print(f'{finding_index=}')
finding_index = lst.index(1) # сохраняет индекс первого совпаденя слева
print(f'{finding_index=}')
# print(lst.index(5)) # попытка вывести индекс значения которого нет -- приведет к ошибке ValueError
```

`[list].sort( reverse = False )` -- Сортирует список `[list]` в порядке возрастания. Если поставить аргумени `reverse=True` то сортировка `[list]` произойдет в порядке убывания. Если внутри списка несколько типов данных, то возникает ошибка, так как нельзя произвести сравнение например между строкой и числом

```python
print(lst.sort()) # выполнили сортировку, и теперь lst у нас сортированный список, несмотря на то что мы его не переприсваивали
print(f'{lst=}') # проверим
lst.sort(reverse=True) # А теперь сортируем в обратном порядке
print(f'{lst=}') # проверим
# print(f'{lst3=}')
# lst3.sort() 
# print(f'{lst3=}')
ls = ['2','a','5','1', 'bb','ab','aa']
ls.sort()
print(f'{ls=}')
```

`[list].reverse()` - Переворачивает элементы списка(зеркалит)

```python
print(lst2)
print(lst2.reverse()) # значения не выдает
print(lst2)
```

`[list].copy()` - создает клон списка `[list]`

```python
lst_new = lst # вроде бы скопировали. | Это не копирование, это ссылка. Т.е. мы ссылаемся в перменной lst_new на lst
# А значит ФИКСИРУЕМ все что происходит в списке lst в список lst_new 
print(f'{lst_new=}\n{lst=}')
lst.append(2333)
print(f'{lst_new=}\n{lst=}')
```

```python
lst_new = lst.copy() # сделали РЕАЛЬНУЮ копию, а не ссылку как в прошлом блоке
print(f'{lst_new=}\n{lst=}')
lst.append(7777)
print(f'{lst_new=}\n{lst=}')
```

## Любое значение в списке можно заменять обращаясь к его индексу с **переприсвоением** значения

```python
print(f'{lst=}')
lst[-1] = 7
print(f'{lst=}')
lst[0] = 5
print(f'{lst=}')
```
