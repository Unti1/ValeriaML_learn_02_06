Индексация - процесс обращения к конкретному объету внутри итерируемова объекта. Любое обращение к индексам работает через `*объект*[*индекс*]`

Итерируемый объект - объект который хранит в себе последовательность объетов (т.е. <u>строка</u> - последовательность букв заключенных в ковычки)

```python
"Hello world"
# Индексация выглядит следующим образом
#      0123456789 10
print("Hello world"[10])
print("Hello world"[0])
# print("Hello world"[11]) # ошибка
```

Индексовый срез - когда мы задаем начальный индекс и конечный индекс через `:`. И чтение идет от начального индекса <u>включительно</u> до конечного <u>исключительно</u>

```python
print("Hello world"[0:5])
print("Hello world"[0:6])
print("Hello world"[0:7])
```

Обратное индексирование - обратный процесс индексации, когда мы идем не с лева-направо, справа-налево. Соответвенно, индекс начинает тоже обратный отсчёт от 0, т.е. -1, -2, ... , -N *длина объекта*



```python
print("STRING"[-6])
print("STRING"[-6:-3])
print("STRING"[-3:-6]) # Ничего не получим тк шаг идет всегда вправо
print("STRING"[-3:-6:-1]) #:-1 - это шаг -1 по срезу. По умолчанию стоит 1  
```

# Форматирование

```python
# через функцию .format()
name = "Боб"
age = 20
text = "Привет {}. Тебе {} лет".format(name, age)
print(text)

text = "Привет {1}. Тебе {0} лет".format(name, age)
print(text)
```

```python
# Вставка в позиции %s 
text = "Привет %s. Тебе %s лет" % (name,age) # работает как format только через % 
print(text)
```

```python
# Вставка в позиции %s 
args = (name,age)
text = "Привет %s. Тебе %s лет" % args # работает как format только через % 
print(text)
```

```python
# Наиболее предпочтительный вариант. F-строка, вставка происходит с уже готовыми переменными
text = f"Привет {name}. Тебе {age} лет"
print(text)
```

`[str].upper()` - ставит все буквы строки `[str]` в верхний регистр (Все буквы заглавные)

```python
print(name)
print(name.upper()) # вот так результат не сохраняется
print(name) 
# Можно сохранить его вручную 
s = name.upper()
#name = name.upper()
print(s)
```

`[str].lower()` - ставит все буквы строки `[str]` в нижний регистр (Все буквы строчные)

```python
print(name)
print(name.lower()) # вот так результат не сохраняется
print(name) 
# Можно сохранить его вручную 
s = name.lower()
print(s)
# или переприсвоить значение
# name = name.lower()
```

`[str].replace(<old>, <new>)` - Заменяет все совпавшие значения в строке `[str]` из подстроки `<old>` на подстроку `<new>`

```python
text = "Какой то там текст"
print(text)
print(text.replace('то ', ''))
print(text.replace('там', 'там там'))
print(text)
text = text.replace('там', 'там там')
print(text)
```

`[str].strip()` - `[str]` очищает от пробельных символов в начале и в конце

```python
str_with_spaces = '                               строка с пробелами                       \n'
print(f'{str_with_spaces.strip()=}')
print(f'{str_with_spaces=}')
str_with_spaces = str_with_spaces.strip()
print(f'{str_with_spaces=}')
```

`[str].split(<разделитель> - по умолчанию пробел, <количество> -  по умолчанию до конца*)` - Разбивает строчку на отдельные части по `<разделителю>` определенное `<количество>` раз

```python
letter = "Пришло письмо с каким-то количеством слов"
print(f"{letter.split()=}")

print(f"{letter.split(maxsplit=3)=}")

print(f"{letter.split(' ',3)=}")
print(f"{letter.split(None,3)=}")

print(f"{letter.split('-')=}")

print(f"{letter.split(sep='-')=}")

letter = "Пришло    письмо           с  каким-то    количеством     слов"

print(f"{letter.split()=}")

words_in_letter = letter.split()

print(words_in_letter, type(words_in_letter))
```

`[str].count(<подстрока>, <начальный индекс, по умол. 0>, <конечный индекс, по умол. -1>)` - Подсчитывает количество `<подстрок>` в строке `[str]` на заданном интервале (`<начальный индекс>`, `<конечный индекс> - 1`)

```python
letter = 'Слово слово слово, за какие то cлова'
print(f"{letter.count('С')=}")

print(f"{letter.count('с')=}")

print(f"{letter.count('слово')=}")

print(f"{letter.count('слово', 7)=}", letter[7:])

print(f"{letter.count('cлова')=}")

print(f"{letter.count('слово', 7, 11)=}", letter[7:11])
```

```python
letter = "Пришло письмо с каким-то количеством слов"
splited_text = letter.split()
print(f'{splited_text=}')
print(f"{ '|'.join('123') =}")
print(f"{ ''.join(splited_text) =}")
print(f"{ '@'.join(splited_text) =}")
print(f"{ ' | '.join(splited_text) =}")
```
